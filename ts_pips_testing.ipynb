```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import random
import ts_pips
```

```python
data_PATH: str = './testing-data-5m.csv'
df = pd.read_csv(data_PATH, header=None)
print(df.head())
```

```python
points = df.drop(columns=[1,2,3])
print(points.head())
```

```python
points = points.to_numpy()
points.shape
```

```python
random_int = random.randint(0, 8100)
ts = points[random_int:random_int+50]
n = 10


xpoints = ts[:, 0]
ypoints = ts[:, 1]
plt.subplot(2, 1, 1)
plt.plot(xpoints, ypoints)
plt.title("Raw time series")

pips_vd = ts_pips.pip_VD(ts, n)
xpoints = pips_vd[:, 0]
ypoints = pips_vd[:, 1]
plt.subplot(2, 1, 2)
plt.plot(xpoints, ypoints)
plt.title("PIP-VD")

plt.show()
```
